```python
# For å jobbe med første del av notatboken trenger vi disse bibliotekene
import math
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
```

```python
# Given values
theta1 = 95     # Vinkel mellom positiv x-akse og "overarm" i grader
l1 = 60        # Lengde til overarm [cm]
theta2 = -100   # Vinkel mellom overarm og underarm i grader
l2 = 50        # Lengde til underarm [cm]
```

```python
# Convert angles to radians
theta1_rad = np.deg2rad(theta1)
theta2_rad = np.deg2rad(theta2)
```

```python
# Original equations
xend = l1 * np.cos(theta1_rad) + l2 * np.cos(theta1_rad + theta2_rad)
yend = l1 * np.sin(theta1_rad) + l2 * np.sin(theta1_rad + theta2_rad)
endepunkt_underarm = [xend, yend]
```

```python
#Analytical solution for partial derivatives. Dette ble også gjort for hånd

# Partial derivatives of xend and yend  with respect to theta1
dx_dtheta1 = -l1 * np.sin(theta1_rad) - l2 * np.sin(theta1_rad + theta2_rad)
# Partial derivative of xend with respect to theta2
dx_dtheta2 = -l2 * np.sin(theta1_rad + theta2_rad)

# Output the partial derivatives at utgangsposisjon-values
print(f"Partial derivative of x with respect to theta1: {dx_dtheta1:.2f}" )
print(f"Partial derivative of x with respect to theta2: {dx_dtheta2:.2f}" )

```

## Numerical solution for partial derivatives    dx/dtheta1 and dx/dtheta2

```python
# Numerical solution for partial derivatives    dx/dtheta1 and dx/dtheta2

# Create list for theta1  ∈ [0, 180°], with 360 elements, the first 180 elements being zero values
theta1_list = np.concatenate((np.zeros(180), np.arange(180, 360)))
#Create list for theta2 ∈ [-180°, 180°], with 360 elements
theta2_list = np.linspace(-180, 180, 360)

print(len(theta1_list))
print(f"theta1_list: {theta1_list}")
```

```python
#dx/dtheta1:

#  THETA2 = konstant -100 degrees (i radianer)
theta2 = (np.pi/180)*(-100)

# Create list for x-values for each theta1
x_koordinat_sluttposisjon_θ_1 = []
for theta1 in theta1_list:
    # Convert angles to radians
    theta1_rad = np.pi/180*theta1
    theta2_konstant = theta2
    # Original equations
    xend = l1 * np.cos(theta1_rad) + l2 * np.cos(theta1_rad + theta2_konstant)
    
    x_koordinat_sluttposisjon_θ_1.append(xend)
```

## Lager grafen ved endringer på  θ_1

```python
#Graph 1:


# make graph with x_koordinat_sluttposisjon as y-values and theta1_list as x-values

plt.plot(theta1_list, x_koordinat_sluttposisjon_θ_1)
plt.xlabel("θ_1 ∈ [0, 180°]")
plt.ylabel("x_koordinat_sluttposisjon")
plt.title("θ_2 er konstant =  -100°")
plt.legend(["Endring på θ_1"])

plt.grid(True)
plt.show()
```

## Grafen for endringer på θ_2

```python
#dx/dθ_2: 

#  Setter θ_1 konstant til 95 grader (i radianer)
theta1 = (np.pi/180)*(95)

x_koordinat_sluttposisjon_θ_2 = []

for theta2 in theta2_list:
    # Convert angles to radians
    theta1_konstant = theta1
    theta2_rad = np.pi/180*theta2
    # Original equations
    xend = l1 * np.cos(theta1_konstant) + l2 * np.cos(theta1_konstant + theta2_rad)
    #yend = l1 * np.sin(theta1_rad) + l2 * np.sin(theta1_rad + theta2_rad)
    x_koordinat_sluttposisjon_θ_2.append(xend)
```

```python
#Graph 2:


# Grafen til endringer i θ_2:
# make graph with x_koordinat_sluttposisjon as y-values and theta1_list as x-values

plt.plot(theta2_list, x_koordinat_sluttposisjon_θ_2)
plt.xlabel("θ_1 er konstant")
plt.ylabel("x_koordinat_sluttposisjon")
plt.title("θ_2 ∈ [-180°, 180°]")
plt.xlim(-180, 180)
plt.grid(True)
plt.show()
```

```python
# Plot both plt.plot(theta1_list, x_koordinat_sluttposisjon) and plt.plot(theta2_list, x_koordinat_sluttposisjon) in the same graph
plt.plot(theta1_list, x_koordinat_sluttposisjon_θ_1, label='Endringer på θ_1')
plt.plot(theta2_list, x_koordinat_sluttposisjon_θ_2, label='Enringer på θ_2')
plt.xlabel("θ_1 ∈ [0, 180°] og θ_2 ∈ [-180°, 180°]")
plt.ylabel("x_koordinat_sluttposisjon")
plt.title("θ_1 og θ_2")
plt.grid(True)
plt.legend()
plt.show()
```
